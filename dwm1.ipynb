```python
print("Characteristics:")
tips.describe(include="category")
```

```python
mean = tips["total_bill"].mean()
median = tips["total_bill"].median()
print(f"The mean bill amount is {mean:.3f}, while the median bill is {median:.3f}")
```

```python
mean = tips["tip"].mean()
median = tips["tip"].median()
print(f"The mean tip amount is {mean:.3f}, while the median tip is {median:.3f}")
```

```python
print(f"The variance and standard deviation bill amount is {tips['total_bill'].var():.3f}, {tips['total_bill'].std():.3f}")
print(f"The variance and standard deviation bill amount is {tips['tip'].var():.3f}, {tips['tip'].std():.3f}")
print(f"The variance and standard deviation bill amount is {tips['size'].var():.3f}, {tips['size'].std():.3f}")
```

```python
min = tips["total_bill"].min()
max = tips["total_bill"].max()
print(f"The minimum bill amount is ${min}, while the maximum bill amount is ${max}, giving range of {max - min}")
min = tips["tip"].min()
max = tips["tip"].max()
print(f"The minimum bill amount is ${min}, while the maximum bill amount is ${max}, giving range of {max - min}")

print(f"Number of people in each dining party varies from {tips['size'].min()} and {tips['size'].max()}")
tips.var()
```

```python
print(f"The median bill amount is ${tips['total_bill'].quantile(q=0.5)}, and the median tip amount is $ {tips['tip'].quantile(q=0.5)}")
print(f"The total bill IQR is the range from {tips['total_bill'].quantile(q=0.25):.2f} to {tips['total_bill'].quantile(q=0.75):.2f}")
print(f"The tip IQR ranges from ${tips['tip'].quantile(q=0.25):.2f} to ${tips['tip'].quantile(q=0.75):.2f}")
```

```python
sns.set(style="ticks", palette="muted")

f, axes = plt.subplots(1, 2, figsize=(12, 4))

day_order=["Thur", "Fri", "Sat", "Sun"]

sns.countplot(x="day", data=tips, hue="time", palette=["teal", "yellow"], order=day_order, ax=axes[0])
axes[0].set_title("Count of tables served by Day")

sns.countplot(x="day", hue="size", data=tips, ax=axes[1], order=day_order)
axes[1].set_title("Tables served by Day and by party size")
```
